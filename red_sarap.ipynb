```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine=pd.read_csv('winequality-red.csv', sep=';')
df=wine.copy()
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
correlation_matris=df.corr()
```

```python
correlation_matris
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matris, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Wine Quality Dataset')
plt.show()
```
